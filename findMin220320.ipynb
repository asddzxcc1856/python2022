```python
# Find min of a parabola y=(w-0.5)^2, how to stop the oscillation in the last part?
```

```python
import numpy as np
```

plot the figure

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ww=np.linspace(-3,3,100)
def f(ww,a,b,c):
    return a*ww**2+b*ww+c;    #a=1, b=-1, c=0.25 for y=(w-0.5)^2
```

```python
import time
plt.plot(ww,f(ww,1,-1,0.25), label='$y=(w-0.5)^{2}$', color='green')
plt.plot(2,f(2,1,-1,0.25), 'bo')
plt.xlabel('w-axis')
plt.ylabel('y-axis')
plt.legend(bbox_to_anchor=(1,0.2))
plt.axis([-3,4,-1,7])
time.sleep(0.4)
plt.plot(0,f(0,1,-1,0.25), 'bo')

plt.show()
```

```python
import pylab as pl
from IPython import display
pl.plot(ww,f(ww,1,1,0.25), label='$y=(w-0.5)^{2}$', color='green')
pl.plot(2,f(2,1,1,0.25), 'bo')
pl.xlabel('w-axis')
pl.ylabel('y-axis')
pl.legend(bbox_to_anchor=(1,0.2))
pl.axis([-3,3,-1,7])
for i in range(4):
#    pl.plot(pl.randn(10))
  #  print(pl.randn(1))
    display.clear_output(wait=True)
    pl.plot(i/2,f(i/2,1,1,0.25), 'bo')
    display.display(pl.gcf())
    time.sleep(0.5)
```

```python
ww=np.linspace(-10.5,11.5,100)
w=np.random.rand(1)-0.5 #find the syntax
w=5
print("w= ", w)
alpha=0.01
# alpha=1     #Will oscillate
dw=1
pl.plot(ww,f(ww,1,-1,0.25), label='$y=(w-0.5)^{2}$', color='green')
# pl.plot(2,f(2,1,1,0.25), 'bo')
pl.xlabel('w-axis')
pl.ylabel('y-axis')
pl.legend(bbox_to_anchor=(1,0.2))
pl.axis([-10.5,11.5,-.1,140])
display.clear_output(wait=True)
i=True
while dw*dw>0.00001:    
    dw=2*(w-0.5)    
    w=w-alpha*dw
    display.clear_output(wait=True)
    if i==True:
        pl.plot(w,f(w,1,-1,0.25), 'bo')
        i=False
    else: 
        pl.plot(w,f(w,1,-1,0.25), 'ro')
        i=True
    display.display(pl.gcf())
    time.sleep(.1)
    print("w=", w, ", y", (w-0.5)**2, ", alpha*dw", alpha*dw)
    time.sleep(.2)

```
